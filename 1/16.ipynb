```python
import random
import matplotlib.pyplot as plt
```

#### Funcion de distribucion

Si x pertenece a [0, a/2]

\begin{equation}
f_x(x) = \frac{2c}{a} x 
\end{equation}

Si x pertenece a (a/2, a]

\begin{equation}
f_x(x) = - \frac{2c}{a} x + 2c
\end{equation}

```python
a = 1
c = 1

# para que sea funcion densidad se tiene que cumplir esto.
def a_(c_):
    return 1/(1-c_)

def f_x(x):
    if x <= a/2:
        return ((2*c)/a) * x
    elif x > (a/2) and x < a:
        return -((2*c)/a) * x + 2*c
    else:
        return 0
```

```python
values_x = []
values_y = []
for i in range(1000):
    x = random.random()
    y = f_x(x)
    values_x.append(x)
    values_y.append(y)
plt.scatter(values_x, values_y, c ="blue")
```
